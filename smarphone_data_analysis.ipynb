```python
# importing the libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
sns.set_style('whitegrid')
```

```python
# importing the excel file containing scraped data

df = pd.read_excel('smartphone_data.xlsx')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
for col in df.columns:
    print(f'{col}: {df[col].unique()}')
    print()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df[['name', 'variant']] = df['name'].str.split('(', n=1, expand=True)
df['name'] = df['name'].str.rstrip()
```

```python
df['brand'] = df['name'].apply(lambda x:x.split()[0])
df['brand'] = df['brand'].apply(lambda x:x.title() if x in ('REDMI','MOTOROLA','SAMSUNG') else x)
df['brand'] = df['brand'].replace("Mi","Xiaomi").replace("I","I Kall")
```

```python
df['colour'] = df['variant'].str.split(', ', n=1, expand=True)[0]
```

```python
df[['sp','mrp']] = df[['sp','mrp']].map(lambda x:x.replace('₹','').replace(',','') if x is not np.nan else np.nan)
```

```python
df['discount'] = df['discount'].apply(lambda x:x.replace('% off','') if x is not np.nan else np.nan)
```

```python
df[['reviews', 'ratings']] = df[['reviews', 'ratings']].map(lambda x:x.replace(',','') if x is not np.nan else np.nan)
```

```python
df['display size'] = df['display size'].str.split('(',  expand=True)[1].apply(lambda x: x.replace(' inch)', '') if x is not None else np.nan)
```

```python
df['processor brand'] = df['processor brand'].replace('Snapdragon','Qualcomm')
```

```python
df['5G ready'] = df['network type'].str.contains('5G')
df['5G ready'] = df['5G ready'].replace({False:'0',True:'1'})
```

```python
df['hybrid sim slot'] = df['hybrid sim slot'].replace({'Yes':'1','No':'0','Dual Sim':'0'})
```

```python
df['bluetooth version'] = df['bluetooth version'].replace('v|V','',regex=True).replace('Android|A2DP',np.NaN,regex=True)
```

```python
df['battery capacity'] = df['battery capacity'].map(lambda x:x.replace(' mAh', '')if x is not np.nan else np.nan)
```

```python
df[['width','height','depth']] = df[['width','height','depth']].map(lambda x:x.replace('mm', '')if x is not np.nan else np.nan)
```

```python
df['weight'] = df['weight'].map(lambda x:x.replace('g', '')if x is not np.nan else np.nan)
```

```python
# dropping columns that don't add much value or contain inconsistent information

df = df.drop(columns=['variant', 'display type', 'resolution', 'wi-fi version', 'operating system','micro usb version'])
```

```python
df.head()
```

```python
# filling na values

num_cols = ['sp','mrp','discount','star','reviews','hybrid sim slot','ratings','display size',
            'battery capacity','width','height','weight','depth','5G ready']
df[num_cols] = df[num_cols].fillna(0)
```

```python
object_cols = [x for x in df.columns if x not in num_cols+['nfc','audio jack']]
df[object_cols] = df[object_cols].fillna("unknown")
```

```python
df.isna().sum()
```

```python
df['nfc'].value_counts(normalize=True)*100
```

```python
df['nfc'] = df['nfc'].fillna('No')
df['nfc'] = df['nfc'].replace({'Yes':'1','No':'0'})
```

```python
df['audio jack'].unique()
```

```python
df['audio jack'].value_counts()
```

```python
replace = ['USB Type-C', 'Type C', 'Type-C', 'USB (Type-C)', 'USB (Type C)', 'USB Type C']

for item in replace:
    df['audio jack'] = df['audio jack'].replace(item,"Type C")
```

```python
replace = ['3.5mm', '3.5mm Headphone Jack', '3.5 mm', '3.5', '3.5MM STEREO']

for item in replace:
    df['audio jack'] = df['audio jack'].replace(item,"3.5mm")
```

```python
replace = ['No', 'Yes', 'YES']

for item in replace:
    df['audio jack'] = df['audio jack'].replace(item,"Type C")
```

```python
df['audio jack'] = df['audio jack'].fillna("unknown")
```

```python
df['audio jack'].value_counts()
```

```python
df.isna().sum()
```

```python
for col in df.columns[9:]:
    print(f'{col}: {df[col].unique()}')
    print()
```

```python
# dropping rows with incorrect data

df = df.drop(index=df[df['width'].isin(['2500 mAh','3000 mAh'])].index)
df = df.drop(df[df['ram']=="32 MB"].index)
```

```python
# changing data types

int_cols = ['sp', 'mrp', 'discount','ratings', 'reviews','5G ready','hybrid sim slot','nfc','battery capacity']
int_dict = {col:'int64' for col in int_cols }
df = df.astype(int_dict)
```

```python
float_cols = ['star','display size','width', 'height', 'depth','weight']
float_dict = {col:'float64' for col in float_cols }
df = df.astype(float_dict)
```

```python
df.info()
```

```python
# detecting outliers

num_data = df.select_dtypes(include=['int64','float64'])

fig, axes = plt.subplots(nrows=num_data.shape[1],ncols=1,figsize=(15, 80))
plt.subplots_adjust(hspace=0.5)

for index,col in enumerate(num_data.columns):
    ax = sns.boxplot(data=num_data, x=col,ax=axes[index])
    ax.set(title=f'Distribution of {col}')
```

```python
# removing outliers

df = df.drop(df.query('height>0 & height<25').index)
df = df.drop(df.query('weight>250 | (weight>0 & weight<150)').index)
df = df.drop(df.query('depth>25').index)
df = df.drop(df[df['display size'] < 5].index)
```

```python
df.shape
```

```python
# columns to be grouped by

groupby_cols = ['name', 'internal storage','ram'] + [x for x in df.columns if x not in ["colour", 'name', 'internal storage','ram']]
```

```python
# grouping smartphones with same name and specifications but different colors into a single row

df = df.groupby(groupby_cols, as_index=False).agg({'colour': ', '.join})
```

```python
# count of internal storage size

sort_order = ['32 GB','64 GB','128 GB','256 GB','512 GB']
ax = sns.countplot(data=df,x='internal storage',order=sort_order)
ax.set(title='Count of internal storages')
plt.show()
```

```python
# average rating for each internal storage size

storage_star = df.groupby('internal storage')['star'].mean().sort_values(ascending=True).to_frame().reset_index()
ax = sns.lineplot(storage_star,x='internal storage',y='star', marker='o')
ax.set(title='Average rating by internal storage',ylabel='average star rating')
plt.show()
```

```python
# count of ram size

sort_order = ['2 GB','3 GB','4 GB','6 GB','8 GB','12 GB','16 GB','unknown']
ax = sns.countplot(df,x='ram',order=sort_order)
ax.set(title='Count of ram size')
plt.show()
```

```python
# average rating for each ram size

ram_star = df.groupby('ram')['star'].mean().sort_values(ascending=True).to_frame().reset_index()

ax = sns.lineplot(ram_star,x='ram',y='star', marker='o')
ax.set(title='Average rating by ram size',ylabel='average star rating')
plt.show()
```

```python
df['sp'].quantile(np.linspace(0.1,1.0,10))
```

```python
# binning the selling price

bins = [0, 10000, 20000, 30000, df['sp'].max()]
labels = [str(val)+'-'+str(bins[idx+1]) for idx,val in enumerate(bins) if idx < len(bins)-2]+["30000+"]

df['sp_range'] = pd.cut(df['sp'],bins=bins, labels=labels)
df['sp_range'].value_counts().sort_index()
```

```python
# average star ratings by selling price range

sp_range_star = df.groupby('sp_range',observed=False)['star'].mean().sort_values(ascending=True).to_frame().reset_index()
ax = sns.lineplot(sp_range_star,x='sp_range',y='star', marker='o')
ax.set(title='Average rating by selling price range',xlabel='selling price range', ylabel='average star rating')
plt.show()
```

```python
# creating price categories by binning sp

avg_sp_brands = df.groupby('brand')['sp'].mean().sort_values(ascending=False).to_frame().reset_index()

bins = [0,15000,20000,30000,avg_sp_brands['sp'].max()]
labels = ['Budget','Mid Range','Upper Mid Range','Premium']

avg_sp_brands['price_category'] = pd.cut(avg_sp_brands['sp'],bins=bins,labels=labels)
```

```python
avg_sp_brands.head()
```

```python
# average selling price of brands

plt.figure(figsize=(10,5))
ax = sns.barplot(avg_sp_brands, y='brand', x='sp',hue='price_category')
ax.set(title='Average selling price of brands',xlabel='selling price',ylabel='')
plt.show()
```

```python
nfc_brand = df.groupby(['brand'])['nfc'].mean().sort_values(ascending=False).to_frame().reset_index().rename(columns={'nfc':'nfc%'})
nfc_brand['nfc%'] = round(nfc_brand['nfc%']*100,0)
nfc_brand = pd.merge(nfc_brand,avg_sp_brands,on='brand').drop('sp',axis=1)
```

```python
# nfc avilability in % by brand

ax = sns.barplot(nfc_brand.drop(nfc_brand[nfc_brand['nfc%']==0].index),
                 y='brand', x='nfc%',hue='price_category')
ax.set(title='nfc avilability in % by brand',xlabel='nfc%', ylabel='')
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```
